Somebody asked at some point if I could turn my image scraping notebook into an installable library, and I've been meaning to have a look at [nbdev](https://nbdev.fast.ai/) for a while anyway.

Nbdev is great as it basically lets you develop a software library in Jupyter notebooks, and then turns them into library + documentation + install packages. I can't recommend it highly enough. It's well worth watching the youtube video to get an idea of what it does.

So now you can use my image scraper like this (which is far more convenient, let's be honest):

```
!pip install -q jmd_imagescraper
```

```
from pathlib import Path
root = Path().cwd()/"images"

from jmd_imagescraper.core import * # dont't worry, it's designed to work with import *

duckduckgo_search(root, "Cats", "cute kittens", max_results=10)
duckduckgo_search(root, "Dogs", "cute puppies", max_results=10)
duckduckgo_search(root, "Birds", "cute baby ducks and chickens", max_results=10)
```

```
from jmd_imagescraper.imagecleaner import *

display_image_cleaner(root)
```

Boom. Docs are at https://joedockrill.github.io/jmd_imagescraper/, the repo it was all generated from is at https://github.com/joedockrill/jmd_imagescraper/ but the meat of it is basically just in 00_core.ipynb, 01_imagecleaner.ipynb and index.ipynb.
